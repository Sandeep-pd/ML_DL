```python
# impute age based on Pclass in 177 records (which contain NaN value)
def fill_age(cols):
  age = cols[0]
  pclass = cols[1]
  if pd.isnull(age):  # 177 records
    if pclass==1:
      return 38
    elif pclass==2:
      return 30
    else:
      return 25
  else:   # 714 records
    return age
```

```python
df['age'] = df[['age','pclass']].apply(fill_age,axis=1)
```

```python
df.info()
```

```python
# Check correlation
df.drop(['sex','who','embark_town'],axis=1).corr() * 100
```

```python
# df.dropna().info()
df.dropna(inplace=True)
```

```python
df = pd.concat([df,pd.get_dummies(df['sex'],drop_first=True),pd.get_dummies(df['who'],drop_first=True),pd.get_dummies(df['embark_town'],drop_first=True)],axis=1)
```

```python
df.info()
```

```python
df.drop(['sex','who','embark_town'],axis=1,inplace=True)
```

```python
df.info()
```

```python
x = df.drop('survived',axis=1)
y = df['survived']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test , y_train,y_test =  train_test_split(x,y,test_size=.2)
```

```python
x_test.shape
```

```python
x_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logr = LogisticRegression()
```

```python
# train the model
logr.fit(x_train,y_train)
```

```python
# predict taget column values
y_pred = logr.predict(x_test)
```

```python
# check the accuracy
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(y_test,y_pred)*100
```

```python
confusion_matrix(y_test,y_pred)
```
