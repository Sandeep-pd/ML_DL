```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df  = pd.read_csv("https://raw.githubusercontent.com/mukulsinghal001/USA-Housing-Price-Prediction/refs/heads/main/USA_Housing.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop("Address",axis=1).corr()*100
```

```python
x=df["Avg. Area Income"]
y= df['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
Ling = LinearRegression()
```

```python
Ling.fit(x_train.values.reshape(-1,1),y_train)
```

```python
y_pred = Ling.predict(x_test.values.reshape(-1,1))
```

```python
plt.scatter(x_test,y_test)
```

```python
#mae,mse,rmse
x_train.shape
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error
```

```python
mean_absolute_error(y_test, y_pred)

```
